```python
import boto3 
import json
import decimal
from boto3.dynamodb.conditions import Key, Attr
from botocore.exceptions import ClientError
%run s3.ipynb

# Helper class to convert a DynamoDB item to JSON.
class DecimalEncoder(json.JSONEncoder):
    def default(self, o):
        if isinstance(o, decimal.Decimal):
            if o % 1 > 0:
                return float(o)
            else:
                return int(o)
        return super(DecimalEncoder, self).default(o)

class user():
    def __init__(self, name):
        self.name=name
        
    def login(self):
        dynamodb = boto3.resource("dynamodb", region_name='us-east-1')
        table = dynamodb.Table('users')
        try:
            response = table.get_item(
                Key={
                    'user': self.name
                }
            )
        except ClientError as e:
            print(e.response['Error']['Message'])
            return False
        else:
            try:
                item = response['Item']
                #print("GetItem succeeded:")
                #print(json.dumps(item, indent=4, cls=DecimalEncoder))
                return True
            except:
                return False
        
    def createUserTable(self):
        dynamodb = boto3.resource('dynamodb', region_name='us-east-1')
        table = dynamodb.create_table(
            TableName=self.name+"projects",
            KeySchema=[
                {
                    'AttributeName': 'project',
                    'KeyType': 'HASH'  #Partition key
                }
            ],
            AttributeDefinitions=[
                {
                    'AttributeName': 'project',
                    'AttributeType': 'S'
                }              

            ],
            ProvisionedThroughput={
                'ReadCapacityUnits': 1,
                'WriteCapacityUnits': 1,
            }
        )    
        print("Table status:", table.table_status)       
        
        
    def register(self):
            ans=self.login()
            if (ans==False):
                s3=s3boy            
                dynamodb = boto3.resource('dynamodb', region_name='us-east-1')            
                table = dynamodb.Table('users')
                table.put_item(
                Item={
                   'user':self.name
                }
                )
                print(self.name+" is now registered!")  

                if(s3.create_bucket("brownboymanbuttxx"+self.name,"us-east-2")==True):
                    print(self.name+"'s bucket has now been created!'")
                    #return True
                else:
                    print("there was problem registering that bucket")

                self.createUserTable()    
                return True
            else:
                print("that name is already taken")
                return False
            
    def createProjectTable(self, pname):
        dynamodb = boto3.resource('dynamodb', region_name='us-east-1')
        strippedpname=pname.replace(" ", "")
        print(strippedpname)
        table = dynamodb.create_table(
            TableName=self.name+"project"+strippedpname,
            KeySchema=[
                {
                    'AttributeName': 'task',
                    'KeyType': 'HASH'  #Partition key
                }
            ],
            AttributeDefinitions=[
                {
                    'AttributeName': 'task',
                    'AttributeType': 'S'
                }
            ],
            ProvisionedThroughput={
                'ReadCapacityUnits': 1,
                'WriteCapacityUnits': 1,
            }
        )
        table = dynamodb.Table(self.name+"projects")
        project=pname
        response = table.put_item(
           Item={
                "project":pname
            }
        )
        print("Project created successfully")
        #print("PutItem succeeded:")
        #print(json.dumps(response, indent=4, cls=DecimalEncoder))

    def listProjects(self):
        dynamodb = boto3.resource('dynamodb', region_name='us-east-1')
        table = dynamodb.Table(self.name+"projects")
        response = table.scan()
        listy=[]
        for i in response['Items']:
            listy.append(i["project"])
        return listy
    
    
    def addTask(self, pname, task, date):
        dynamodb = boto3.resource('dynamodb', region_name='us-east-1')
        strippedpname=pname.replace(" ", "")
        table = dynamodb.Table(self.name+"project"+strippedpname)
        response = table.put_item(
           Item={
                "task":task,
                "duedate":date,
                "completed":"N"
            })
        print("task "+task+" added successfully")
    
    def listTasks(self, pname):
        dynamodb = boto3.resource('dynamodb', region_name='us-east-1')
        strippedpname=pname.replace(" ", "")
        table = dynamodb.Table(self.name+"project"+strippedpname)
        listy=[]
        try:
            response = table.scan()
        except:
            return listy
        
        x=0
        for i in response['Items']:
            x=x+1
            listy.append(i["task"] )
            
            print(str(x)+". "+i["task"]+ " - " +i["duedate"] + " - "+i["completed"])
    
        return listy
    
    def completeTask(self, pname, task):
        dynamodb = boto3.resource('dynamodb', region_name='us-east-1')
        strippedpname=pname.replace(" ", "")
        table = dynamodb.Table(self.name+"project"+strippedpname)
        response = table.update_item(
            Key={
                'task': task
            },
            UpdateExpression="set completed = :c",
            ExpressionAttributeValues={
                ':c': "Y"
            },
            ReturnValues="UPDATED_NEW"
        )
```
